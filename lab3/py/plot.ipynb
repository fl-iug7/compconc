# Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Funções

```python
def A(ts, tp):
    return np.array(ts) / np.array(tp)

def E(A_values):
    return A_values / 6
```

# Plots

## Matriz (500x500)

```python
ts = [0.448264]
tp = [0.540947, 0.270841, 0.156186, 0.089466]
t_values = [1, 2, 4, 8]
A_values = A(ts, tp)


plt.plot(t_values, A_values, marker='o', label='A(n, t)')
plt.xlabel('Número de threads')
plt.ylabel('A(n, t)')
plt.title('Plot de A(n, t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
E_values = E(A_values)

plt.plot(t_values, E_values, marker='o', label='E(n, t)')
plt.xlabel('Número de threads')
plt.ylabel('E(n, t)')
plt.title('Plot de E(n, t)')
plt.legend()
plt.grid(True)
plt.show()
```
